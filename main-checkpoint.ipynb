```python
import cv2
import numpy as np
from pygame import mixer
```

```python
import torch
import torch.nn as nn
from torchvision.models import mobilenet_v2
from torch.nn import functional as F

class Eye_model(nn.Module):
    def __init__(self, num_classes=2):
        super(Eye_model, self).__init__()

        # Load pretrained MobilenetV2 model
        self.base_model = mobilenet_v2(pretrained=True)
        out_features = self.base_model.classifier[1].out_features
        
        # Add custom layers
        self.head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(out_features, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, num_classes),
        )
        
        # Freeze base model layers
        for param in self.base_model.parameters():
            param.requires_grad = False

    def forward(self, inp):
        x = self.base_model(inp)
        if isinstance(x, tuple):
            x = x[0]
        return self.head(x)

model = Eye_model(num_classes=2)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Load weights
model.load_state_dict(torch.load(r"D:\CSES\Year2\IoT\Project\checkpoint.pt", map_location=torch.device('cpu')))
```

```python
mixer.init()
sound= mixer.Sound(r'D:\CSES\Year2\IoT\Project\alarm.wav')
cap = cv2.VideoCapture(0)
Score = 0
while True:
    ret, frame = cap.read()
    height,width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces= face_cascade.detectMultiScale(gray, scaleFactor= 1.2, minNeighbors=3)
    eyes= eye_cascade.detectMultiScale(gray, scaleFactor= 1.1, minNeighbors=1)
    
    cv2.rectangle(frame, (0,height-50),(200,height),(0,0,0),thickness=cv2.FILLED)
    
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,pt1=(x,y),pt2=(x+w,y+h), color= (255,0,0), thickness=3 )
        
    for (ex,ey,ew,eh) in eyes:
        #cv2.rectangle(frame,pt1=(ex,ey),pt2=(ex+ew,ey+eh), color= (255,0,0), thickness=3 )
        
        # preprocessing steps
        eye = frame[ey:ey+eh,ex:ex+ew]
        eye = cv2.resize(eye,(80,80))
        eye = eye / 255.0
        eye = np.transpose(eye, (2, 0, 1))
        eye = torch.from_numpy(eye).float()
        eye = eye.unsqueeze(0)

        # preprocessing is done now model prediction
        model.eval()
        with torch.no_grad():
            prediction = model(eye)
            prediction = torch.sigmoid(prediction).numpy()
        
        # if eyes are closed
        if prediction[0][0] > 0.30:
            cv2.putText(frame,'closed',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)
            cv2.putText(frame,'Score'+str(Score),(100,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)
            Score = Score+1
            if(Score > 7):
                try:
                    sound.play()
                except:
                    pass
            
        # if eyes are open
        elif prediction[0][1]>0.90:
            cv2.putText(frame,'open',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)      
            cv2.putText(frame,'Score'+str(Score),(100,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)
            Score = max(0, Score-1)
        
    cv2.imshow('frame',frame)
    if cv2.waitKey(33) & 0xFF==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "D:\CSES\Year2\IoT\Project\pngtree-eye-of-love-png-image_4669618.jpg"  # Replace with your image path
image = cv2.imread(image_path)  # Reads image in BGR format

# Convert to grayscale for eye detection
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect eyes using Haar Cascade (ensure eye_cascade is loaded)
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)
```

```python
id = 0
for (ex, ey, ew, eh) in eyes:
    # Extract eye region
    eye_region = image[ey:ey+eh, ex:ex+ew]
    id += 1
    plt.subplot(1, 3, id)
    plt.imshow(eye_region)

    
    # Preprocess for PyTorch model (resize, normalize, reshape)
    eye_processed = cv2.resize(eye_region, (80, 80))  # Resize to model's expected input
    eye_processed = eye_processed / 255.0  # Normalize to [0, 1]
    eye_processed = np.transpose(eye_processed, (2, 0, 1))  # Change from HWC to CHW format
    eye_tensor = torch.from_numpy(eye_processed).float()  # Convert to PyTorch tensor
    eye_tensor = eye_tensor.unsqueeze(0)  # Add batch dimension [1, 3, 80, 80]

    # Ensure model is in evaluation mode
    model.eval()

    # Disable gradients for inference
    with torch.no_grad():
        prediction = model(eye_tensor)  # Forward pass
        prediction_prob = torch.sigmoid(prediction).numpy()  # Convert logits to probabilities

    print(prediction_prob[0][0], prediction_prob)

    # Interpret prediction (adjust thresholds as needed)
    if prediction_prob[0][0] > 0.30:  # Assuming output is [prob_closed, prob_open]
        print("Eye State: Closed")
    else:
        print("Eye State: Open")
```
